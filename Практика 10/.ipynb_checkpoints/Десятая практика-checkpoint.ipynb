```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
stat_dpi = 90
big_pics = (20, 20)
long_pics = (20, 10)
```

### Модель Броуновского движения и ее графики

```python
def plot_bm(N = 600, x_0=0, y_0=0):
    broun_points = np.array([x_0, y_0])
    
    def _broun_moving(data, N):
        nonlocal broun_points
        if N==0:
            return None
        data_new = data + np.random.normal(0, 1, 2)
        broun_points = np.vstack((broun_points, data_new))
        return _broun_moving(data_new, N-1)
        
    _broun_moving(broun_points, N)
    
    plt.figure(facecolor="white", figsize = long_pics, dpi=stat_dpi)
    plt.plot(broun_points[:,0], linewidth=2)
    plt.plot(broun_points[:,1], linewidth=2)
    plt.show()
    
    plt.figure(facecolor="white", figsize = big_pics, dpi=stat_dpi)
#     for i in range(1, broun_points.shape[0]):
#         plt.plot(broun_points[i-1, :], broun_points[i, :], '-')
    plt.plot(broun_points[:, 0], broun_points[:, 1], '-r')
    plt.plot(broun_points[:, 0], broun_points[:, 1], 'ob')
    return broun_points, N, x_0, y_0
```

### Алгоритм Ричардсона v.1
Дельту по x принимаем как константу равную разности первого и последнего x

```python
def richardv1(data, delta = 300, N = 1300):
    
    delta_x = abs(data[-1, 0] - data[0, 0])
    x=[]
    y=[]
    
    for j in range(1,delta):
        m = int(N/j)
        #print(f'{m=}, {data[N]=}, {data[i*m]=}, \n{(data[N]-data[i*m])**2=}')
        r = (np.sum((data[N, :]-data[j*m, :])**2))**0.5
        delta_x_j = delta_x*j/N
        
        L_temp=0
        
        
        for n in range(2, m):
            L_temp += ( (data[n*j,1] - data[(n-1)*j, 1])**2 + delta_x_j**2 )**0.5
            
        L_temp += r
        
    
        x.append(np.log(delta_x_j))
        y.append(np.log(L_temp))
    
    plt.figure(facecolor="white", figsize = long_pics, dpi=stat_dpi)   
    x = np.array(x)
    y = np.array(y)
    plt.scatter(x, y, s=30) 

    model = np.polyfit(x, y, 1)
    print(f"Fractal degree = {1-model[0]}")
    p = np.poly1d(model)
    plt.plot(x,p(x),"r-")
        
        
        
```

### Алгоритм Ричардсона v.2
Дельту по x высчитываем внутри итераций с 2 до m, в качестве результирующей берем среднеквадратическую

```python
def richardv2(data, delta = 300, N = 600):
    
    x=[]
    y=[]
    
    for j in range(1,delta):
        m = int(N/j)
        #print(f'{m=}, {data[N]=}, {data[i*m]=}, \n{(data[N]-data[i*m])**2=}')
        r = (np.sum((data[N, :]-data[j*m, :])**2))**0.5
        
        L_temp = 0
        delta_x_temp = 0
        
        for n in range(2, m):
            dx = (data[n*j,0] - data[(n-1)*j, 0])**2
            L_temp += ( (data[n*j,1] - data[(n-1)*j, 1])**2 + dx )**0.5
            delta_x_temp+=dx
        
            
        L_temp += r
        
    
        x.append( np.log( (delta_x_temp)**0.5/m ) )
        y.append(np.log(L_temp))
    
    plt.figure(facecolor="white", figsize = long_pics, dpi=stat_dpi)    
    x = np.array(x)
    y = np.array(y)
    plt.scatter(x, y, s=30) 

    model = np.polyfit(x, y, 1)
    print(f"Fractal degree = {1-model[0]}")
    p = np.poly1d(model)
    plt.plot(x,p(x),"r-")
```

### Алгоритм Фёдера (сеточный метод)

```python
def eps_f(x, n):
    return int(len(x)/n)
    

def NN(n, data):
    m = 100
    M = 100
    x = data[:, 0]
    y = data[:, 1]
    x1 = x[0]
    eps = eps_f(x, n)
    for i in range(n-1):
        xx_i = x1 + eps*i
        xx_i1 = x1+eps*(i+1)
        delta = (xx_i1-xx_i)/M
        
        d = [y[int(xx_i+delta*j)] for j in range(M)]
        A = (np.max(d) - np.min(d))/eps
        m = m + np.ceil(A)
        
    return m, eps
    
def find_d1(data, start, end):
    NN_end, eps_end = NN(end, data)
    NN_start, eps_start = NN(start, data)
    print(NN_start)
    print(NN_end)
    print(eps_start)
    print(eps_end)
    D1 = ( np.log(NN_end)-np.log(NN_start) )/( np.log(eps_start) - np.log(eps_end))
    return D1


    
```

```python
dset = plot_bm(N=1300, x_0=0.5, y_0=1)
```

```python
richardv1(data=dset[0], delta = 300, N=dset[1])
```

```python
richardv2(data=dset[0], delta = 200, N=dset[1])
```

```python
test_data = []
for n in range(20, 300, 10):
    y, x = NN(n, dset[0])
    test_data.append([x, y])
    
plt.figure(facecolor="white", figsize = long_pics, dpi=stat_dpi) 
test_data = np.log(np.array(test_data))
x = test_data[:, 0]
y = test_data[:, 1]
plt.scatter(x, y, color="black")

model = np.polyfit(x, y, 1)
print(f"Fractal degree = {1-model[0]}")
p = np.poly1d(model)
plt.plot(x,p(x),"r-")
```
