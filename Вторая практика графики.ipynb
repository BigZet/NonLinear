```python
import scipy.integrate as integr            
import numpy as np                          
import matplotlib.pyplot as plt 
import math


colors = ["#00FFFF",                
          "#0000FF",                
          "#FF0000",                
          "#00FF00",                
          "#D2691E",                
          "#AAA662",
          "#9A0EEA",
          "#808000",
          "#FFA500",
          "#9ACD32",
          "#ED0DD9",
          "#FC5A50",
          "#929591",
          "#029386",
          "#C79FEF",
          "#FAC205",
          "#6E750E",
          "#06C2AC",
          "#CD5C5C",                
          "#A52A2A",                
          "#7B68EE",                
          "#4682B4",                
          "#800000"                
          ]
colors = colors*3

def abline(slope, intercept):
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, c = "black")
    
def mesh(x, y):
    general = []
    for i in range(len(x)):
        for j in range(len(y)):
            local = []
            local.append(x[i])
            local.append(y[j])
            general.append(local)
    return general

stat_dpi = 120
rect_pics = (7, 7)
dynamics_pics = (14, 7)
```

Первый номер

```python
def f(y, t, epsilon = 1):
    return [-epsilon*math.sin(y)]

time = np.linspace(0, 5, 100)
fig = plt.figure(facecolor="white", figsize = dynamics_pics, dpi=stat_dpi)
grid = mesh([0, math.pi, 1, math.pi-0.1], [1])

for i in grid:
    w = integr.odeint(f,                   
                  i[0],           
                  time,         
                  (i[1], )         
                  )

    y = w[:, 0]
    plt.plot(time, y, c = "r" if i[1]>=0 else "b", label = "x(time)")
    plt.scatter(0, i[0])
    
plt.grid(True)  
plt.show() 
```

Третий номер

```python
def f(y, t, epsilon = 1):
    y1 = epsilon*y-y**2
    #print(y1)
    return [y1[0]]
time = np.linspace(0, 5, 100)
fig = plt.figure(facecolor="white", figsize = dynamics_pics, dpi=stat_dpi)
grid = mesh([0.1, 0, 0.5, math.pi, 1], [2])

for i in grid:
    w = integr.odeint(f,                   
                  i[0],           
                  time,         
                  (i[1], )         
                  )

    y = w[:, 0]
    plt.plot(time, y, c = "r" if i[1]>=0 else "b", label = "x(time)")
    plt.scatter(0, i[0])
    
plt.grid(True)  
plt.show() 
```

Седьмой номер

```python
def f(y, t, epsilon = 1):
    y1, y2 = y
    y1t = epsilon*y1-y2-y1*(y2**2+y1**2)
    y2t = epsilon*y2+y1-y2*(y2**2+y1**2)
    return [y1t, y2t]

time = np.linspace(0, 5, 100)
fig = plt.figure(facecolor="white", figsize = dynamics_pics, dpi=stat_dpi)
grid = mesh([-2, -1, 0, 1, 2], [-2, -1, 0, 1, 2])
grid_p = mesh(grid, [1, 2])

for i in grid_p:
    w = integr.odeint(f,                   
                  i[0],           
                  time,         
                  (i[1], )         
                  )

    y1 = w[:, 0]
    y2 = w[:, 1]
    plt.plot(y1, y2, c = colors[i[1]+i[0][1]+i[0][0]], label = "x(time)")
    plt.scatter(y1[0], y2[0], c = colors[i[1]+i[0][1]+i[0][0]])
    
plt.grid(True)  
plt.show() 
```

Пятый номер

```python
def f(y, t, epsilon = 1):
    y1, y2 = y
    y1t = y2
    y2t = -4*y1**3-2*epsilon*y1
    return [y1t, y2t]

time = np.linspace(0, 5, 100)
fig = plt.figure(facecolor="white", figsize = dynamics_pics, dpi=stat_dpi)
grid = mesh([-2, -1, 0, 1, 2], [-2, -1, 0, 1, 2])
grid_p = mesh(grid, [1, 2])

for i in grid_p:
    w = integr.odeint(f,                   
                  i[0],           
                  time,         
                  (i[1], )         
                  )

    y1 = w[:, 0]
    y2 = w[:, 1]
    plt.plot(y1, y2, c = colors[i[1]+i[0][1]+i[0][0]], label = "x(time)")
    plt.scatter(y1[0], y2[0], c = colors[i[1]+i[0][1]+i[0][0]])
    
plt.grid(True)  
plt.show() 
```

Шестой номер

```python
def f(y, t, epsilon = 1):
    y1, y2 = y
    y1t = epsilon*y1
    y2t = -y2
    return [y1t, y2t]

time = np.linspace(0, 5, 100)
fig = plt.figure(facecolor="white", figsize = dynamics_pics, dpi=stat_dpi)
grid = mesh([-1, 0, 1], [-1, 0, 1])
grid_p = mesh(grid, [1, -1])

for i in grid_p:
    w = integr.odeint(f,                   
                  i[0],           
                  time,         
                  (i[1], )         
                  )

    y1 = w[:, 0]
    y2 = w[:, 1]
    plt.plot(time, y2, c = colors[i[1]+i[0][1]+i[0][0]], label = "x(time)")
    plt.plot(time, y1, c = colors[i[1]+i[0][1]+i[0][0]], label = "x(time)")
    plt.scatter(y1[0], y2[0], c = colors[i[1]+i[0][1]+i[0][0]])
    
plt.grid(True)  
plt.show() 
```
